```python
# File paths for quick adjustment if places or directories change.
path_input = '00_data_input'
path_temp = '01_data_temp'
path_inter = '02_data_intermediate'
path_output = '03_data_output'
path_obesity = os.path.join(path_inter, 'obesity_refined_data.csv')
path_poverty = os.path.join(path_inter, 'poverty_refined_data.csv')
```

**Read dataframes** and combine obesity and poverty refined raw data.

```python
df_obesity = pd.read_csv(path_obesity)
df_poverty = pd.read_csv(path_poverty)
```

```python
df_obese_poor = df_obesity[['county_fips','county','state','PCT_obese_adults_2013']]
df_obese_poor = df_obese_poor.merge(df_poverty[['county_fips','PCT_poverty_rate_2015']], how='inner', on='county_fips')
df_obese_poor[:4]
```

### Relational Graphs 

```python
sns.lmplot(x='PCT_obese_adults_2013', y='PCT_poverty_rate_2015', data=df_obese_poor, size=10)
```

```python
sns.jointplot(x='PCT_obese_adults_2013', y='PCT_poverty_rate_2015', data=df_obese_poor, kind='hex', size=14)
```
