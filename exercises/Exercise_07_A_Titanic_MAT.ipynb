## Naive Bayes

```python
count_total = df_train.index.size
count_survived = 0
count_survived_male = 0
count_survived_female = 0
count_male = 0

for row in range(count_total):
    isMale = (df_train.iloc[row]['Sex'] == 'male')
    if (isMale):
        count_male = count_male + 1
    if (df_train.iloc[row]['Survived'] == 1):   
        count_survived = count_survived + 1
        if (isMale):
            count_survived_male = count_survived_male + 1
            
count_survived_female = count_survived - count_survived_male
count_female = count_total - count_male

df_gender = pd.DataFrame(index=['total','male','female'],columns=['survived','passengers'],
                         data=np.array([[count_survived,count_survived_male,count_survived_female],
                                       [count_total, count_male, count_female]]).transpose())

probabilities = [df_gender['survived']['total']/df_gender['passengers']['total'],
                 df_gender['survived']['male']/df_gender['passengers']['male'],
                 df_gender['survived']['female']/df_gender['passengers']['female']]

df_gender['probabilities']=probabilities
df_gender
```

```python
def naive_bayes(feature):
    p = df_gender['probabilities'][feature] * (df_gender['passengers'][feature]/df_gender['passengers']['total']) / \
                df_gender['probabilities']['total']
    return p
    
print(naive_bayes('male'))
print(naive_bayes('female'))
```
